```python
import matplotlib.pyplot as plt
import numpy as np
import pickle 

%matplotlib inline
```

```python
input_filename = "results.pkl"
read_mode = "rb"
```

```python
with open(input_filename, read_mode) as inputfile:
    data = pickle.load(inputfile)
```

```python
data
```

```python
random_data = np.array(data["random"])
unc_data = np.array(data["uncertainty"])
pts = unc_data.shape[1]
steps = np.arange(100, (pts + 1) * 100, 100)
```

```python
mean = random_data.mean(axis=0)
std = random_data.std(axis=0)

plt.plot(steps, mean)
plt.fill_between(steps, mean-std, mean+std, alpha=0.3, label="Random")

mean = unc_data.mean(axis=0)
std = unc_data.std(axis=0)
plt.plot(steps, mean)
plt.fill_between(steps, mean-std, mean+std, alpha=0.3, label="Uncertainty")

plt.grid(alpha=0.2)
plt.ylim([0.75, 0.95])
plt.xlabel("Points Added")
plt.ylabel("Test Accuracy")
plt.legend(frameon=False)
plt.savefig("compare.pdf", bbox_inches="tight")
plt.savefig("compare.png", bbox_inches="tight")
```

```python
mean = unc_data.mean(axis=0)
std = unc_data.std(axis=0)

plt.plot(steps, mean)
plt.fill_between(steps, mean-std, mean+std, alpha=0.3)
plt.grid(alpha=0.2)
plt.ylim([0.0, 1.0])
plt.xlabel("Points Added")
plt.ylabel("Test Accuracy")
```
